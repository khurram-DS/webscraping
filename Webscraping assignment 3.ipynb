# Qno-2 In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a dataframe and csv. In case if any product vertical has

less than 3 pages in search results then scrape all the products available under that product

vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of

Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"

and “Product URL”. In case, if any of the details are missing for any of the product then

replace it by “NA“.


```python
# Function Definition
def amzon_speaker(url):
    driver=webdriver.Chrome(r"C:\Users\khurr\Downloads\chromedriver.exe")
#calling empty list first to store the data
    Product_link=[]
    product_name=[]
    no_of_ratings=[]
    expected_delivery=[]
    availability=[]
    Return_or_Exchange=[]
    other=[]
    price=[]
    Ratings=[]
#navigating url to scrap the mazon data
    driver.get(url)
#giving input as a speaker in search bar
    search_field_product=driver.find_element_by_id("twotabsearchtextbox")
    search_field_product.send_keys("speaker")
    search_button=driver.find_element_by_xpath("//div[@class='nav-right']/div/span/input")
    search_button.click()
# lets extract all the product link till page 4
    for i in range(0,3):
        url_tags=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
        for i in url_tags:
            Product_link.append(i.get_attribute('href'))
    driver.find_element_by_xpath("//li[@class='a-last']").click()
    time.sleep(3)
# after getting the each product link lets navigate and extract the individual data    
    for j in Product_link:
        driver.get(j) 
# extractig product name.
        try:
            product_name.append((driver.find_element_by_xpath("//h1[@class='a-size-large a-spacing-none']/span")).text)
        except NoSuchElementException:
            product_name.append("NA")
# extracting number of ratings
        try:
            no_of_ratings.append((driver.find_element_by_xpath("//div[@class='a-row a-spacing-medium averageStarRatingNumerical']/span")).text)
        except NoSuchElementException:
            no_of_ratings.append("NA")
# extracting expected delivery time
        try:
            expected_delivery.append((driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini a-spacing-top-mini']/div")).text)
        except NoSuchElementException:
            expected_delivery.append("NA")
# extracting availability of the product
        try:
            availability.append((driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")).text)
        except NoSuchElementException:
            availability.append("NA")
# extracting exchange or return data
        try:
            Return_or_Exchange.append((driver.find_element_by_xpath("//div[@class='a-column a-span3 a-text-center icon-container  icon-grid-width']//div[@class='a-section a-spacing-none  icon-content']/a")).text)
        except NoSuchElementException:
            Return_or_Exchange.append("NA")
# extracting other details from the page
        try:
            other.append((driver.find_element_by_xpath("//div[@class='a-section a-spacing-small']/div")).text)
        except NoSuchElementException:
            other.append("NA")
#extracting price of the data    
        try:
            price.append((driver.find_element_by_xpath("//td[@class='a-span12']/span")).text)
        except NoSuchElementException:
            price.append("NA")
#extrcating rating of all the data
        try:
            Ratings.append((driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']/span")).text)
        except NoSuchElementException:
            Ratings.append("NA")
#before dumping the data in dtaa frame let me check the length.        
    print('Length of the list are ==>>',len(product_name),len(availability),len(no_of_ratings),len(Product_link),len(Ratings),len(Return_or_Exchange),len(other),len(expected_delivery),len(price))
# lets make a data frame and save it for further use
    amazon_speaker=pd.DataFrame({'product_name':product_name,
                           'Availability':availability,
                            'No_of_ratings':no_of_ratings,
                            'Ratings':Ratings,
                            'Return/Exchange':Return_or_Exchange,    
                            'Other Details':other, 
                            'Expected_delivery':expected_delivery,
                             'Price':price,
                            'Product_link':Product_link,})
    print(amazon_speaker)
    amazon_speaker.to_csv('amazon_speaker.csv', index = False)
    
    
# Calling Function
amzon_speaker('https://www.amazon.in/')
```

# 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.


```python
def cordinates(url):
    driver=webdriver.Chrome(r"C:\Users\khurr\Downloads\chromedriver.exe")
    city=[]
    driver.get(url)
    search_field_items=driver.find_element_by_id("searchboxinput")
    search_field_items.send_keys("kolkata")
    search_button=driver.find_element_by_xpath("//button[@class='searchbox-searchbutton']")
    search_button.click()
    time.sleep(3)
    cordinates=driver.current_url
    latitude=cordinates.split('@')[1].split('/data')[0].split(',')[0]
    longitude=cordinates.split('@')[1].split('/data')[0].split(',')[1]
    time.sleep(5)
    loc=driver.find_element_by_xpath("//h1[@class='section-hero-header-title-title GLOBAL__gm2-headline-5']/span")
    city.append(loc.text)
    print('Geospatial cordinates of the City ===>>',city)
    print('Latitude of your location is ===>> ' ,latitude)
    print('Longitude of your location is ===>> ' ,longitude)
# Calling Function
cordinates('https://www.google.com/maps/')    
```

# 6. Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 –
September 20) from trak.in.


```python
def trak(url):
    driver=webdriver.Chrome(r"C:\Users\khurr\Downloads\chromedriver.exe")
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    tables = soup.find_all("table")
 #july data
    table = tables[6]
    tab_data = [[cell.text for cell in row.find_all(["th","td"])]
                            for row in table.find_all("tr")]
    july = pd.DataFrame(tab_data)
 #droping the unneccesary columns which of no use
    july.columns = july.iloc[0,:]
    july.drop(index=0,inplace=True)
    july.set_index('Sr. No.',inplace=True)
    print(july)
    
#august data
    table = tables[4]
    tab_data = [[cell.text for cell in row.find_all(["th","td"])]
                            for row in table.find_all("tr")]
    aug = pd.DataFrame(tab_data)
#droping the unneccesary columns which of no use
    aug.columns = aug.iloc[0,:]
    aug.drop(index=0,inplace=True)
    aug.set_index('Sr. No.',inplace=True)
    print(aug)
    
#september data
    table = tables[5]
    tab_data = [[cell.text for cell in row.find_all(["th","td"])]
                            for row in table.find_all("tr")]
    sep = pd.DataFrame(tab_data)
#droping the unneccesary columns which of no use
    sep.columns = sep.iloc[0,:]
    sep.drop(index=0,inplace=True)
    sep.set_index('Sr. No.',inplace=True)
    print(sep)
    
    
    
# Calling Function
trak('https://trak.in/india-startup-funding-investment-2015/')    
```

# 7. Write a program to scrap all the available details of top 10 gaming laptops from digit.in. 

```python
driver=webdriver.Chrome(r"C:\Users\khurr\Downloads\chromedriver.exe")
```

```python
driver.get('https://www.digit.in/')
```

```python
#clickng on top 10 option 
top_10=driver.find_element_by_xpath("//div[@class='menu']/ul/li[3]/a")
top_10.click()

time.sleep(3)
#clicking on laptops option
laptops=driver.find_element_by_xpath("//div[@class='categoty_list']//button[2][@class='tablinks ']")
laptops.click()


#best gaming laptops link
best_gaming=driver.find_element_by_xpath("//div[@id='laptops']//div[3][@class='thumb_cont']/a/div")
best_gaming.click()

#intialising lists
name = []
price = []
OS = []
display = []
processor = []
HDD = []
RAM = []
weight = []
dimension = []
GPU = []

#names
names=driver.find_elements_by_xpath("//div[@class='right-container']/div/a/h3")
for i in names:
    name.append(i.text)
    
#os
os=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[1]/div/div")
for i in os:
    OS.append(i.text)
    
#display
displays=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[2]/div/div")
for i in displays:
    display.append(i.text)
    
#processor
processors=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[3]/div/div")
for i in processors:
    processor.append(i.text)
processor

#memory
memories=driver.find_elements_by_xpath("//div[@class='Spcs-details'][1]/table/tbody/tr[6]/td[1]")#list of specificaion name
memories_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details'][1]/table/tbody/tr[6]/td[3]")#values of specifiations 
for i in range(len(memories)):
        if memories[i].text=='Memory':
            HDD.append(memories_spec[i].text.split('/')[0])
            RAM.append(memories_spec[i].text.split('/')[1])
        else:
            HDD.append('No details available')#append no details as value for memory is missing in some of the laptops
            RAM.append('No details available')#append no details as value for memory is missing in some of the laptops

#weight
weights=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[1]")#list of specificaion name
weight_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[3]")#values of specifiations
for i in range(len(weights)):
        if weights[i].text=='Weight':
            weight.append(weight_spec[i].text)
        
#dimension
dimension=[]
dims=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[1]")#list of specificaion name
dims_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[3]")#values of specifiations
for i in range(len(dims)):
        if dims[i].text=='Dimension':
            dimension.append(dims_spec[i].text)

#graphical processor
GPUs=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[1]")#list of specificaion name
GPUs_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[3]")#values of specifiations
for i in range(len(GPUs)):
        if GPUs[i].text=='Graphics Processor':
            GPU.append(GPUs_spec[i].text)
        
full_specs=[]
urls=driver.find_elements_by_xpath("//div[@class='full-specs']/span")#getting the url of full specs links
for i in urls:
    if i.get_attribute('data-href'):
        full_specs.append(i.get_attribute('data-href'))
    
for i in full_specs:#iterating throug every laptops full specs' page
    driver.get(i)
    try:
        prices=driver.find_element_by_xpath("//div[@class='Block-price']/b")
        price.append(prices.text)
    except NoSuchElementException:#exception handling for no price details
        price.append("No details available")
        
df=pd.DataFrame({"Name":name,
                "Price":price,
                "OS":OS,
                "Display":display,
                "HDD":HDD,
                 "RAM":RAM,
                "processor":processor,
                "weight":weight,
                "Dimension":dimension,
                "Graphical processor":GPU})
df
```
